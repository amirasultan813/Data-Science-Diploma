## Wuzzuf Data Science Jobs Scrapping
- You are required to scrap all data science jobs from wuzzuf and save the results in csv file
- the required fields are
<ul>
    <li>job title</li>
    <li>job link</li>
    <li>Location</li>
    <li>skills required </li>
    <li>company_name</li>
</ul>

```python
import requests 
from bs4 import BeautifulSoup
import csv
```

```python
url = 'https://wuzzuf.net/search/jobs/?a=navbl&q=data%20science%20jobs&start=0'
print(url)
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
soup
```

```python
container = soup.find('div' , attrs={'class':'css-9i2afk'})
container
```

```python
rows = container.find_all('div',attrs={'class':'css-1gatmva e1v1l3u10'})
rows
```

## only one job

- Job title

```python
job = rows[0].find('h2',attrs={'class':'css-m604qf'}).find('a').text
job
```

- Job Link

```python
job_link = rows[0].find('h2',attrs={'class':'css-m604qf'}).find('a')['href']
print(job_link)
```

- Location

```python
location = rows[0].find('span',attrs={'class':'css-5wys0k'}).text
location
```

- Company Name

```python
comp_name = rows[0].find('div',attrs={'class':'css-d7j1kk'}).find('a').text
comp_name
```

- Skills Required

```python
skills = rows[0].find('div',attrs={'class':'css-y4udm8'}).find('div',attrs=None).text
print(skills)
```

## All jobs in page

```python
for row in rows:
    
    job = row.find('h2',attrs={'class':'css-m604qf'}).find('a').text
    print(job)
    
    job_link = row.find('h2',attrs={'class':'css-m604qf'}).find('a')['href']
    print(job_link)
    
    location = row.find('span',attrs={'class':'css-5wys0k'}).text
    print(location)
    
    comp_name = row.find('div',attrs={'class':'css-d7j1kk'}).find('a').text
    print(comp_name)
    
    skills = row.find('div',attrs={'class':'css-y4udm8'}).find('div',attrs=None).text
    print(skills)
    
    print('*'*40)
    #break
```

## All jobs in all pages

```python
print(url)
```

```python
job_ , job_lnk , loc , comp , skills_ = [],[],[],[],[]
for pge in range(1,55):
    
    url = f'https://wuzzuf.net/search/jobs/?a=navbl&q=data%20science%20jobs&start={pge}'
    req = requests.get(url)
    
    print(f'_______________________________________{pge}___________________________________________')
    
    soup = BeautifulSoup(req.content, 'html.parser')
    
    container = soup.find('div' , attrs={'class':'css-9i2afk'})
    rows = container.find_all('div',attrs={'class':'css-1gatmva e1v1l3u10'})
    
    for row in rows:
    
        job = row.find('h2',attrs={'class':'css-m604qf'}).find('a').text
        job_.append(job)
        print(job)

        job_link = row.find('h2',attrs={'class':'css-m604qf'}).find('a')['href']
        job_lnk.append(job_link)
        print(job_link)

        location = row.find('span',attrs={'class':'css-5wys0k'}).text
        loc.append(location)
        print(location)

        comp_name = row.find('div',attrs={'class':'css-d7j1kk'}).find('a').text
        comp.append(comp_name)
        print(comp_name)

        skills = row.find('div',attrs={'class':'css-y4udm8'}).find('div',attrs=None).text
        skills_.append(skills)
        print(skills)

        print('*'*40)
    
    #break
```

## Saving into csv file

```python
with open('Wuzuf.csv' , mode='w', newline='', encoding='utf-8')as f:
    
    writer = csv.writer(f)
    writer.writerow(['Job Title' , 'Job link' ,'Company name' , 'Location','Skills'])
    
    for job , joblnk , loc_ , comp_ , skill  in zip(job_ , job_lnk , loc , comp , skills_):
        
        writer.writerow([job , joblnk , loc_ , comp_ , skill])
```
