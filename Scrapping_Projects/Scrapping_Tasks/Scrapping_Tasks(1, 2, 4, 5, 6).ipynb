Task 1
---------
- Scrap the USD To EGP Exchange rate from this website
https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html
and then use it to make a software that takes amount of USD Dollars from the user and calculate how much will it cost in EGP.

```python
url = 'https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html'
req = requests.get(url)
```

```python
req
```

```python
soup = BeautifulSoup(req.content , 'html.parser')
soup
```

```python
container = soup.find('div',attrs={'class':'row'})
```

```python
rate = container.find('div',attrs={'class':'large-5 medium-5 small-12 columns nolpad'}).find('span',attrs={'id':'shd2a'}).text
rate
```

```python
rate_ =  rate.split(' ')
rate_
```

```python
rate_usd = float(rate_[3])
rate_usd
```

```python
def converting(amount , rate):
    
    amount = amount.split(' ')
    value =  float(amount[0])
    res = value * rate if amount[1].upper() == 'USD' else 0
    return res

input_ = (input('Enter amount of USD Dollars : ')).upper()
result = converting(input_ , rate_usd)
print(f'{input_} = {result} EGP') if result != 0 else print('Invalid Currency!')
```

Task 2
--------- 
- Scraping from forecast weather
Imagine you work for forecast weather Now imagine your boss wants you to find the temperture for each day in celsius degree. How could you do this with Beautiful Soup?
https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168

```python
url2 = 'https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168'
req_2 = requests.get(url2)
```

```python
req_2
```

```python
soup2 = BeautifulSoup(req_2.content , 'html.parser')
#soup2
```

```python
container = soup2.find('div',attrs={'id':'seven-day-forecast'})
```

```python
rows = container.find('ul',attrs={'id':'seven-day-forecast-list'}).find_all('li')
```

```python
day_,temp_f,temp_c, high_low = [],[],[],[]
for row in rows:
    
    day = row.find('p',attrs={'class':'period-name'}).text
    day_.append(day)
    
    tempf = row.select_one('p.temp.temp-high, p.temp.temp-low').text
    tempff = (int(tempf.split(' ')[1]))
    
    h_l = (tempf.split(' ')[0])[:-1]
    #(fahrenheit - 32) * 5/9
    tempc = (int(tempf.split(' ')[1])-23) * 5//9
    
    temp_f.append(tempff)
    temp_c.append(tempc)
    high_low.append(h_l)
    
    print(day)
    print(tempff)
    print(tempc)
    print(h_l)
    print('*'*25)
```

```python
print(temp_f)
print(temp_c)
print(day_)
print(high_low)
```

```python
with open('temp.csv' , mode = 'w' , newline='' , encoding = 'utf-8') as f:
    
    writer = csv.writer(f)
    writer.writerow(['Day','High_Low','Temp_F','Temp_C'])
    
    for day,h_l,tempf,tempc in zip(day_,high_low,temp_f,temp_c):
        
        writer.writerow([day,h_l,tempf,tempc])
        
```

```python
data = pd.read_csv('temp.csv')
data
```

Task 4

- once you decided to do some exploring an want to know each country's (name, capital, population, Area(km^2) 
scrape the 250 countries (name, capital, population, area) and save them into a csv file
use this site to scrape this data : https://www.scrapethissite.com/pages/simple/

```python
url4 = 'https://www.scrapethissite.com/pages/simple/'
req_4 = requests.get(url4)
req_4
```

```python
soup_4 = BeautifulSoup(req_4.content , 'html.parser')
#soup_4
```

```python
container4 = soup_4.find('section',attrs={'id':'countries'})
```

```python
rows = container4.find_all('div',attrs={'class':'col-md-4 country'})
rows
```

```python
country_,capital_,population_,area_ = [],[],[],[]
for row in rows:
    
    country = row.find('h3',attrs={'class':'country-name'}).text.strip()
    capital = row.find('span',attrs={'class':'country-capital'}).text
    population = int(row.find('span',attrs={'class':'country-population'}).text)
    area = float(row.find('span',attrs={'class':'country-area'}).text)
    
    country_.append(country)
    capital_.append(capital)
    population_.append(population)
    area_.append(area)
    
    print(country)
    print(capital)
    print(population)
    print(area)
    print('*'*30)
    
```

```python
with open('countries.csv', mode='w' ,newline='' ,encoding='utf-8')as f:
    
    writer = csv.writer(f)
    writer.writerow(['Country','Capital','Population','Area in km^2'])
    
    for con,cap,pop,area in zip(country_,capital_,population_,area_):
        
        writer.writerow([con,cap,pop,area])
```

```python
data_ = pd.read_csv('countries.csv')
data_
```

Task 5
---------
- scraping data about hockey teams:
you have to scrap the first page table's data (Team name, year, wins, losses, win%, goals for(gf), golaagainst(ga)
from this website: https://www.scrapethissite.com/pages/forms/?page_num=1

## Scrapping All pages

```python
name_ , year_ ,wins_ ,losses_ ,win_percent_, goals_ ,golaa_ = [],[],[],[],[],[],[]
for pge in range(1,25):
    
    print(f'______________________________________________{pge}______________________________________________________')
    
    url5 = f'https://www.scrapethissite.com/pages/forms/?page_num={pge}'
    req_5 = requests.get(url5)
    soup_5 = BeautifulSoup(req_5.content, 'html.parser')
    
    table = soup_5.find('table',attrs={'class':'table'})
    trs = table.find_all('tr',attrs={'class':'team'})
    
    for tr in trs:
    
        name = tr.find('td',attrs={'class':'name'}).text.strip()
        year = tr.find('td',attrs={'class':'year'}).text.strip()
        wins = tr.find('td',attrs={'class':'wins'}).text.strip()
        losses = tr.find('td',attrs={'class':'losses'}).text.strip()
        #win_percent = tr.find('td',attrs={'class':'pct text-success'}).text.strip()
        goals = tr.find('td',attrs={'class':'gf'}).text.strip()
        golaa = tr.find('td',attrs={'class':'ga'}).text.strip()

        name_.append(name)
        year_.append(year)
        wins_.append(wins)
        losses_.append(losses)
        #win_percent_.append(win_percent)
        goals_.append(goals)
        golaa_.append(golaa)

        print(name)
        print(year)
        print(wins)
        print(losses)
        #print(win_percent)
        print(goals)
        print(golaa)
        print('*'*30)    
    
```

```python
with open('Hockey_Teams.csv' , mode='w' , newline='' , encoding='utf-8')as f:
    
    writer = csv.writer(f)
    writer.writerow(['Team name', 'Year', 'Wins', 'Losses', 'Goals', 'Golaagainst'])
    
    for nam,year,wins,loss,go,gola in zip(name_ , year_ ,wins_ ,losses_ , goals_ ,golaa_):
        
        writer.writerow([nam,year,wins,loss,go,gola])
```

```python
hockey_data = pd.read_csv('Hockey_Teams.csv')
hockey_data
```

Task 6
---------
- Use this web service 
https://jsonplaceholder.typicode.com/users
and make a program that get all users data including name, username, email, street, suite, city, zip code, geo_lat & geo_long and then save them in a CSV & Excel file.

```python
response = requests.get('https://jsonplaceholder.typicode.com/users')
users_data = response.json()
users_data
```

```python
with open('User_data', 'w', newline='',encoding='utf-8') as f:
    
    writer = csv.DictWriter(f, fieldnames=['Name', 'User_name', 'E-mail', 'Street','Suite','City','Zip_code','Geo_lat','Geo_long'])
    writer.writeheader()
    for user in users_data:
        name = user['name']
        username = user['username']
        email = user['email']
        street = user['address']['street']
        suite = user['address']['suite']
        city = user['address']['city']
        zipcode = user['address']['zipcode']
        geolat = user['address']['geo']['lat']
        geolong = user['address']['geo']['lng']
        
        writer.writerow({'Name':name , 'User_name':username, 'E-mail':email, 'Street':street,'Suite':suite,'City':city,'Zip_code':zipcode,'Geo_lat':geolat,'Geo_long':geolong})
        
```
