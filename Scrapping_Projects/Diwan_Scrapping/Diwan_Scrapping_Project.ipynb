### Import Helper Libraries 

```python
import requests
from bs4 import BeautifulSoup
```

### Request and get HTML page content and parse it 

```python
url = 'https://diwanegypt.com/product-category/books/english-books/'
print(url)
```

```python
req = requests.get(url)
req.content
```

```python
soup = BeautifulSoup(req.content , 'html.parser')
soup
```

### Get One book data 
* name 
* author 
* price 

```python
container = soup.find('div' , attrs={'class':'section-inner products-overview-with-sidebar'})
container
```

```python
ul = container.find('ul',attrs={'class':'products columns-5'})
ul
```

```python
all_ = ul.find_all('li')
all_
```

```python
all_.find('h2').text
```

```python
book = ul.find('li').find('h2').text
book
```

```python
author = ul.find('li').find('span').text
author
```

```python
price = ul.find('li').find('span',attrs={'class':"woocommerce-Price-amount amount"}).text
price
```

```python
ul.find('li').find('bdi').text
```
