```python
import csv 
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://ocw.mit.edu/courses/electrical-engineering-and-computer-science/6-006-introduction-to-algorithms-spring-2020/lecture-videos/#'
```

```python
req = requests.get(url)
req
```

```python
req.content
```

```python
soup = BeautifulSoup(req.content , 'html.parser')
soup
```

```python
container = soup.find('div',attrs={'class':'mb-5'})
cards = container.find_all('div',attrs={'class':'resource-item resource-list-page'})
```

```python
cards[0].find('a' , attrs={'class':"resource-list-title"}).attrs['href']
```

```python
titles , links = [],[]
for crd in cards:
    
    anchor = crd.find('a' , attrs={'class':"resource-list-title"})
    full_link = 'https://ocw.mit.edu/'+ anchor.attrs['href']
    title = anchor.text
    
    links.append(full_link)
    titles.append(title)
```

### Two Tasks together

```python
titles , links ,description , instructors = [],[],[],[]
for crd in cards:
    
    anchor = crd.find('a' , attrs={'class':"resource-list-title"})
    full_link = 'https://ocw.mit.edu/'+ anchor.attrs['href']
    title = anchor.text
    
    links.append(full_link)
    titles.append(title)
    
    req1 = requests.get(full_link)
    con1 = req1.content
    soup1 = BeautifulSoup(con1 , 'html.parser')
    
    desc = soup1.find('div', attrs={'class':'description'})
    p1 = desc.find('p').text.strip()
    p2 = desc.find_all('p')[1].text.split(':')[1]
    
    description.append(p1)
    instructors.append(p2)
    
    print(title)
    print(full_link)
    print(p1)
    print(p2)
    
    print('*'*100)
```

```python
with open('Lec.csv',mode='w',newline='',encoding='utf-8')as f:
    
    writer = csv.writer(f)
    writer.writerow([
        'Lec_Name',
        'Description',
        'Instructor',
        'Link'
    ])
    
    for name,desc,ins,link in zip(titles ,description , instructors  , links):
        
        writer.writerow([
            name,
            desc,
            ins,
            link
        ])
```
